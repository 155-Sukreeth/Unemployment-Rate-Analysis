# Data Analysis

# 1) The Dot-com bubble (1997-2001): Rapid growth of the internet and technology industries

```python
prob1 = df[['1996', '1997', '1998', '1999', '2000','2001', '2002']].copy()
change_before_p1 = ["NA"]*len(prob1)
change_after_p1 = ["NA"]*len(prob1)
prob1['change_before'] = (prob1['1996']-prob1['1997'])/ prob1['1997']
prob1['change_after'] = (prob1['2001']-prob1['2002'])/prob1['2002']
prob1.head()
```

```python
prob1_before = prob1.groupby(prob1.index)['change_before'].max().sort_values(ascending = False)
selected_countries = list(prob1_before.index)
prob1_before = prob1.loc[prob1.index.isin(selected_countries[0:10])]
plt.figure(figsize = (15, 5))
sns.barplot(data = prob1_before, x = prob1_before.index, y = 'change_before').set(title = 'Countries with Highest decrease in Unemployement Rate due to growth in Tech')
prob1_before
```

```python
duration = ['1997', '1998', '1999', '2000','2001']
prob1_selected = {'country_name': [], 'year': [], 'urate': []}
for i in selected_countries[0:10]:
    for j in duration:
        prob1_selected['country_name'].append(i)
        prob1_selected['year'].append(j)
        prob1_selected['urate'].append(prob1_before.loc[i,j])
temp_prob1 = pd.DataFrame(prob1_selected)
plt.figure(figsize=(15,10))
sns.lineplot(data = temp_prob1, hue = 'country_name', x = 'year', y = 'urate').set_title('Unemployment rate during the period')
```

```python
prob1_after = prob1.groupby(prob1.index)['change_after'].max().sort_values()
selected_countries = list(prob1_after.index)
prob1_after = prob1.loc[prob1.index.isin(selected_countries[0:10])]
plt.figure(figsize = (15, 5))
sns.barplot(data = prob1_after, x = prob1_after.index, y = 'change_after').set(title = 'Countries with Highest increase in Unemployement Rate due to Layoffs')
prob1_after
```

```python
prob1_after = prob1.groupby(prob1.index)['change_after'].min().sort_values(ascending = False)
selected_countries = list(prob1_after.index)
prob1_after = prob1.loc[prob1.index.isin(selected_countries[0:10])]
plt.figure(figsize = (20, 5))
sns.barplot(data = prob1_after, x = prob1_after.index, y = 'change_after').set(title = 'Countries whose Unemployement rate dropped despite Layoffs')
prob1_after
```

```python
prob1_mean = [df['1997'].mean(), df['1998'].mean(), df['1999'].mean(), df['2000'].mean(), df['2001'].mean()]
# len(duration), len(prob1_mean)
prob1_avg = pd.DataFrame({'duration': duration, 'prob1_mean': prob1_mean})
sns.lineplot(data = prob1_avg, x = duration, y = prob1_mean).set_title('Average Global Unemployement Rate during the period')
```

We can observe that the raise in technology does not have much affect globally, this suggests that most countries were still relying on non-tech based jobs

# 6) Other Comparisons

Based on GDP, for the past decade

```python
selected_countries = ['United States', 'Hong Kong SAR, China', 'Japan', 'Germany', 'United Kingdom', 'India']
duration = ['2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021']
temp = df.loc[df.index.isin(selected_countries), df.columns.isin(duration)].copy()
temp
```

```python
temp.plot.bar(figsize = (15, 5))
plt.legend(bbox_to_anchor = (1.25, 0.6), loc='center right')
plt.tight_layout()
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(data = temp)
```

Above plot suggests that from the years 2015-2019, these top GDP countries remained consistent. It can be seen that some countries recovered from GFC 2008 due to the raise of AI and other related advancements in technologies especially Japan and China. Pandemic again stretched our density as some countries were affected by covid significantly. Longer violin plots suggests that our countries' unemployement rates are varying significantly from each other.

```python
sns.heatmap(data = temp)
```

From the above and below plots we can infer that: 

India remained most consisitent in mainitaining its rate until the pandemic, it quickly recovered to reach its base unemployement rate.

We can also notice that the temperature of germany decreasing, suggesting that despite crisis' and pandemics it was able to increase their employement opportunities over time.

China maintained low unemployement rate throught out the decade even being the most populated country, we can notice that both India and China were consisitent despite being the two most populous countries in the world.

Japan and China can be seen making the most out of automation industries and AI.

Finally we can observe that despite being on top of the GDP list, US' unemployement rates are quite high, this suggests that a minor job sector of the country is contributing much to its GDP
